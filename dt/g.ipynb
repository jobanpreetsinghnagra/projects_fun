```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
def entropy(p1):
    p0 = 1-p1
    a = -p1*np.log2(p1)
    b = p0*np.log(p0)
    H = a - b
    return H
```

```python
x = np.linspace(0.0,1.0,100)
y = entropy(x)

plt.plot(x,y)
plt.grid()
plt.title("entropy graph")
plt.xlabel("p1")
plt.ylabel("H(p1)")
plt.show()
```
