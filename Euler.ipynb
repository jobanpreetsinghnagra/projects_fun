```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
def euler(i,x):
    ans= math.cos(x) + i*math.sin(x)
    return ans
```

```python
pi_1 = np.arange(0,2*math.pi,0.001,dtype=float)
pi_1
```

```python
pi_2 = np.vectorize(euler) 
"""
Define a vectorized function which takes a nested sequence 
of objects or numpy arrays as inputs and returns a single numpy array or a tuple of numpy arrays. 
The vectorized function evaluates pyfunc over successive tuples of the input arrays like the python map function,
except it uses the broadcasting rules of numpy."""

# values of pi at x , and euler identity at y
plt.plot(pi_1,pi_2(0,pi_1))
plt.show()
```
