```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

## Various Activation Functions
 ### Sigmoid , ReLU , Linear , Softmax

```python
def sigmoid(x):
    ez = 1 + np.exp(-x)
    sigma = 1 / ez
    return sigma
```

```python
def RELU(x):
    r = np.max(0,x)
    return r
```

```python
def linear(x):
    return x
```

```python
def softmax(z):
    exp_z = np.exp(z - np.max(z))  # stability
    return exp_z/ np.sum(exp_z)
```

```python
random.seed(5)
x = np.linspace(-10,10,10000)
y = sigmoid(x)


plt.figure()
fig, ax = plt.subplots()
ax.spines['left'].set_position('zero')  # Move left spine (Y-axis) to x=0
ax.spines['bottom'].set_position('zero') # Move bottom spine (X-axis) to y=0
plt.title("Sigmoid Function")
plt.plot(x,y)
plt.show()
```

```python
random.seed(5)
x = np.linspace(-10,10,10000)
y = linear(x)


plt.figure()
fig, ax = plt.subplots()
ax.spines['left'].set_position('zero')  # Move left spine (Y-axis) to x=0
ax.spines['bottom'].set_position('zero') # Move bottom spine (X-axis) to y=0
plt.title("Linear")
plt.plot(x,y)
plt.show()
```

```python

x = np.linspace(-10, 10, 10)
y1 = []
y2 = []
y3 = []

for val in x:
    z = np.array([val, 0, 0])  # vary only first logit
    s = softmax(z)
    y1.append(s[0])
    y2.append(s[1])
    y3.append(s[2])

# Plot
fig, ax = plt.subplots()
ax.spines['left'].set_position('zero')
ax.spines['bottom'].set_position('zero')
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.plot(x, y1, label='Softmax(z1)')
plt.plot(x, y2, label='Softmax(z2)')
plt.plot(x, y3, label='Softmax(z3)')
plt.title("Softmax Function (3-class example)")
plt.legend()
plt.grid(True, linestyle='--', alpha=0.5)
plt.show()
```
